# Data Preprocessing

```python
def MakeDataset(dir,lab):
    df3=pd.DataFrame(columns=['a1','a2','a3','a4'])

    for root, dirs, f in os.walk(dir):
        for i in range (len(f)):

            path3 = os.path.join(root,f[i])

            df = pd.read_csv(path3)
            load_col = [int(f[i][5:-4])/100 for j in range(len(df))]
            label_col = [lab for j in range(len(df))]
            df['load']=load_col
            df['fault']=label_col

            df3 = pd.concat([df3,df],axis=0)
           

            print(path3)
            
    return df3 
```

MakeDataset is a function which adds load column,fault column to each file and merges entire csv files in a folder to a single data frame

```python
dir='C:\\Users\\91978\\OneDrive\\Desktop\\notes\\BrokenTooth'
df_F = MakeDataset(dir,lab='F')
```

```python
from scipy.stats import chi2
def var_test(x, va0, direction = "two-tailed", alpha = 0.05):
    n = len(x)
    Q = (n - 1) * np.var(x) / va0 
    q1 = chi2.ppf(alpha / 2, n - 1)
    q2 = chi2.ppf(1 - (alpha / 2), n - 1)
    if Q<= q1 or Q>= q2:
        return "H_0 rejected"
    else:
        return "H_0 not rejected"
```

```python
df=pd.read_csv('h30hz0.csv')
var_test(x, va0 = statistics.variance(df['a1']))
if(var_test(x, va0 = statistics.variance(df['a1'])*0.316=="H_0 not rejected"): 
    print("It may be healthy tooth")
else:
    print("It is a broken tooth")
print(var_test(x, va0 = statistics.variance(df['a1'])*0.316))
```
